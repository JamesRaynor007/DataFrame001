```python
import numpy as np
import pandas as pd

# Set seed for reproducibility
np.random.seed(42)

# Generate date range for the year 2025
dates = pd.date_range(start='2025-01-01', end='2025-12-31')

# Generate 10,000 random dates within the range
date_samples = np.random.choice(dates, size=10000, replace=True)

# Generate 100 unique Client IDs
client_ids = np.arange(1, 101)

# List of 100 client names provided
names = [
    "Alejandro", "Isabel", "Carlos", "María", "Diego", "Lucia", "Javier", "Carmen", "Miguel", "Sofia",
    "Pablo", "Victoria", "Antonio", "Elena", "Fernando", "Rosa", "Luis", "Alicia", "José", "Marta",
    "Sergio", "Paula", "Ricardo", "Clara", "Andrés", "Laura", "Manuel", "Ana", "Gabriel", "Teresa",
    "Juan", "Carmen", "Victor", "Natalia", "Rafael", "Inés", "Ernesto", "Beatriz", "César", "Adriana",
    "Ramón", "Silvia", "Oscar", "Dolores", "Guillermo", "Alicia", "Jorge", "Lucia", "Eduardo", "Marcela",
    "Martín", "Paula", "Samuel", "Verónica", "Guillermo", "Carmen", "Sebastián", "Andrea", "Alberto", "Carmen",
    "Diego", "Rosa", "Mario", "Elena", "Víctor", "Isabel", "Daniel", "Mariana", "Nicolás", "Laura",
    "Raúl", "Carmen", "Pedro", "Victoria", "Esteban", "Sonia", "Felipe", "Alicia", "Sergio", "Juana",
    "Javier", "Clara", "Antonio", "Rosa", "Juan Carlos", "Isabel", "José Luis", "María José", "Guillermo", "Ana María",
    "Héctor", "Patricia", "Diego", "Carmen", "Luis Miguel", "Adriana", "Santiago", "Lucia", "Guillermo", "Beatriz"
]

# Map client IDs to names
client_name_map = dict(zip(client_ids, names))

# Assign each of the 10,000 sales to a random client ID
client_id_samples = np.random.choice(client_ids, size=10000, replace=True)

# Map client IDs to names
client_name_samples = [client_name_map[id_] for id_ in client_id_samples]

# Generate random sales amounts between 100 and 1000
sales_amounts = np.round(np.random.uniform(100, 1000, size=10000), 2)

# Create the DataFrame
df3 = pd.DataFrame({
    'Date': date_samples,
    'ID_Client': client_id_samples,
    'Client_Name': client_name_samples,
    'Sales_Amount': sales_amounts
})

# Optional: Shuffle the DataFrame for added randomness
df3 = df3.sample(frac=1).reset_index(drop=True)

# Display the first few rows
print(df3.head())
```

```python
df3.info()
```

```python
# Add a 'Quarter' column based on the 'Date' column
df3['Quarter'] = df3['Date'].dt.quarter.apply(lambda q: f'Q{q}')
# Add 'Week Number' column
df3['Week_Number'] = df3['Date'].dt.isocalendar().week

# Add 'Month' column
df3['Month'] = df3['Date'].dt.strftime('%B')
```
