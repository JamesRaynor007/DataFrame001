```python
import pandas as pd
import numpy as np
import random
from faker import Faker

# Initialize Faker for generating fake names
fake = Faker()

# Set seed for reproducibility
np.random.seed(42)

# Number of rows
num_rows = 1000

# Generate random dates between 01-01-2024 and 12-31-2025
start_date = pd.to_datetime('2024-01-01')
end_date = pd.to_datetime('2025-12-31')
date_range = (end_date - start_date).days

dates = [start_date + pd.Timedelta(days=np.random.randint(0, date_range + 1)) for _ in range(num_rows)]
dates_str = [date.strftime('%m-%d-%Y') for date in dates]

# Extract month name
months = [date.strftime('%B') for date in dates]

# Determine quarter
def get_quarter(date):
    quarter = (date.month - 1) // 3 + 1
    return f'Q{quarter}'

quarters = [get_quarter(date) for date in dates]

# Quarter Year
quarter_years = [f'{q}-{date.year}' for q, date in zip(quarters, dates)]

# Id_Client (max 100 unique)
client_ids = np.random.choice(range(1, 101), size=num_rows, replace=True)
client_ids_unique = np.unique(client_ids)
# To ensure non-repetitive IDs for each row, assign unique IDs
# Let's generate 100 unique IDs and assign randomly
client_id_list = np.random.choice(range(1, 101), size=num_rows, replace=True)
client_name_map = {cid: fake.name() for cid in range(1, 101)}
client_names = [client_name_map[cid] for cid in client_id_list]

# Id_Sale (max 5000 unique)
sale_ids = np.random.choice(range(1, 5001), size=num_rows, replace=False)

# Id_Product (max 1000 unique)
product_ids = np.random.choice(range(1, 1001), size=num_rows, replace=False)

# Product_Name (max 1000 unique)
product_id_to_name = {pid: fake.word().title() for pid in range(1, 1001)}
product_names = [product_id_to_name[pid] for pid in product_ids]

# Id_Product_Type (assume 10 types)
product_types = [f'Type_{i}' for i in range(1, 11)]
product_type_assignments = np.random.choice(product_types, size=num_rows)

# Total_Amount Float (simulate sales amount between 10 and 1000)
total_amounts = np.round(np.random.uniform(10, 1000, size=num_rows), 2)

# Let's fill in columns 12-18 with plausible data:
# 12 - Sales_Channel (String)
sales_channels = ['Online', 'In-Store', 'Partner', 'Wholesale']
sales_channel_choices = np.random.choice(sales_channels, size=num_rows)

# 13 - Discount Applied (Float between 0% and 20%)
discounts = np.round(np.random.uniform(0, 0.2, size=num_rows), 2)

# 14 - Payment_Method (String)
payment_methods = ['Credit Card', 'Debit Card', 'PayPal', 'Bank Transfer']
payment_choices = np.random.choice(payment_methods, size=num_rows)

# 15 - Shipping_Cost (Float)
shipping_costs = np.round(np.random.uniform(5, 50, size=num_rows), 2)

# 16 - Customer_Rating (Integer 1-5)
ratings = np.random.randint(1, 6, size=num_rows)

# 17 - Order_Status (String)
order_statuses = ['Pending', 'Completed', 'Cancelled', 'Shipped']
status_choices = np.random.choice(order_statuses, size=num_rows)

# 18 - Comment (String)
comments = [fake.sentence(nb_words=6) for _ in range(num_rows)]

# Create DataFrame
df = pd.DataFrame({
    'Date': dates_str,
    'Month': months,
    'Quarter': quarters,
    'Quarter_Year': quarter_years,
    'Id_Client': client_id_list,
    'Client_Name': client_names,
    'Id_Sale': sale_ids,
    'Id_Product': product_ids,
    'Product_Name': product_names,
    'Id_Product_Type': product_type_assignments,
    'Total_Amount': total_amounts,
    'Sales_Channel': sales_channel_choices,
    'Discount_Applied': discounts,
    'Payment_Method': payment_choices,
    'Shipping_Cost': shipping_costs,
    'Customer_Rating': ratings,
    'Order_Status': status_choices,
    'Comments': comments
})

# Display the first few rows
print(df.head())
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Set seed for reproducibility
np.random.seed(42)

# Helper functions
def random_date(start_date, end_date):
    """Generate a random date between start_date and end_date."""
    delta = end_date - start_date
    random_days = np.random.randint(0, delta.days + 1)
    return start_date + timedelta(days=random_days)

# Date range
start_dt = datetime(2024, 1, 1)
end_dt = datetime(2025, 12, 31)

# Generate 10000 rows as an example
n_rows = 10000

# 1. Date
dates = [random_date(start_dt, end_dt) for _ in range(n_rows)]
dates_str = [dt.strftime('%m-%d-%Y') for dt in dates]

# 2. Month name
months = [dt.strftime('%B') for dt in dates]

# 3. Quarter
def get_quarter(date):
    month = date.month
    if month in [1,2,3]:
        return 'Q1'
    elif month in [4,5,6]:
        return 'Q2'
    elif month in [7,8,9]:
        return 'Q3'
    else:
        return 'Q4'

quarters = [get_quarter(dt) for dt in dates]

# 4. Quarter Year
quarter_years = [f"{q}-{dt.year}" for q, dt in zip(quarters, dates)]

# 5. Id_Client (max 100 unique)
client_ids = np.arange(1, 101)
id_clients = np.random.choice(client_ids, size=n_rows)

# 6. Client_Name
client_names = [f'Client_{cid}' for cid in id_clients]

# 7. Id_Sale (max 5000 unique)
sale_ids = np.arange(1, 5001)
id_sales = np.random.choice(sale_ids, size=n_rows)

# 8. Id_Product (max 1000 unique)
product_ids = np.arange(1, 1001)
id_products = np.random.choice(product_ids, size=n_rows)

# 9. Product_Name
product_names = [f'Product_{pid}' for pid in id_products]

# 10. Id_Product_Type (max 50 unique)
product_type_ids = np.arange(1, 51)
id_product_types = np.random.choice(product_type_ids, size=n_rows)

# 11. Units
units = np.random.randint(1, 101, size=n_rows)  # 1 to 100 units

# 12. Unit_Cost
unit_costs = np.round(np.random.uniform(5, 500, size=n_rows), 2)

# 13. Margin_Applied (1% to 50%)
margins = np.round(np.random.uniform(0.01, 0.50, size=n_rows), 2)

# 14. Unit_Price (calculated)
unit_prices = np.round(unit_costs * (1 + margins), 2)

# 15. Total_Amount (Units * Unit_Price)
total_amounts = np.round(units * unit_prices, 2)

# 16. Shipping Method (max 3)
shipping_methods = ['Standard', 'Express', 'Overnight']
shipping_choices = np.random.choice(shipping_methods, size=n_rows)

# 17. Payment Status (max 3)
payment_statuses = ['Transfer', 'Cash', 'Credit Card']
payment_choices = np.random.choice(payment_statuses, size=n_rows)

# 18. Shipping Date (random between date and date + 10 days)
shipping_dates = []
for dt in dates:
    shipping_dt = dt + timedelta(days=np.random.randint(0, 11))
    shipping_dates.append(shipping_dt.strftime('%m-%d-%Y'))

# Construct the DataFrame
df = pd.DataFrame({
    'Date': dates_str,
    'Month': months,
    'Quarter': quarters,
    'Quarter_Year': quarter_years,
    'Id_Client': id_clients,
    'Client_Name': client_names,
    'Id_Sale': id_sales,
    'Id_Product': id_products,
    'Product_Name': product_names,
    'Id_Product_Type': id_product_types,
    'Units': units,
    'Unit_Cost': unit_costs,
    'Margin_Applied': margins,
    'Unit_Price': unit_prices,
    'Total_Amount': total_amounts,
    'Shipping_Method': shipping_choices,
    'Payment_Status': payment_choices,
    'Shipping_Date': shipping_dates
})

# Display the first few rows
print(df.head())

# Optionally, save to CSV
# df.to_csv('random_dataframe.csv', index=False)
```
